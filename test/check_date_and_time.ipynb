```python
import pandas as pd
```

```python
y = '2022'
m = '10'
d = '30'
time_from = '01:30'
time_to = '03:30'

df = pd.read_csv(f'./files/{y}-{m}/{y}-{m}-{d}.csv')
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df = df[df['Time'].dt.microsecond == 0]
df.reset_index(drop=True, inplace=True)
df.head()
```

#### Change time format

```python
fin = 'Europe/Helsinki'
swe = 'Europe/Stockholm'
df['Time'] = pd.to_datetime(df['Time']).dt.tz_localize(fin, ambiguous='infer')
df['Time'] = df['Time'].dt.tz_convert(swe)
```

```python
start_time = f'{y}-{m}-{d} {time_from}:00'
end_time = f'{y}-{m}-{d} {time_to}:00'
start_time = pd.to_datetime(start_time).tz_localize(swe)
end_time = pd.to_datetime(end_time).tz_localize(swe)
filtered_data_time = df[(df['Time'] >= start_time) & (df['Time'] <= end_time)]
pd.set_option('display.max_rows', None)
print(filtered_data_time)
```
